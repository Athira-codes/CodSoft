```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/DATASETS/LoanApprovalPrediction.csv')
df
```

```python
df.drop(['Loan_ID','Education'],axis=1,inplace=True)
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['Credit_History'].unique()
```

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mode()[0])
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mode()[0])
df
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
le1=LabelEncoder()
df['Married']=le1.fit_transform(df['Married'])
le2=LabelEncoder()
df['Dependents']=le2.fit_transform(df['Dependents'])
le3=LabelEncoder()
df['Self_Employed']=le3.fit_transform(df['Self_Employed'])
le4=LabelEncoder()
df['Property_Area']=le4.fit_transform(df['Property_Area'])
le5=LabelEncoder()
df['Loan_Status']=le5.fit_transform(df['Loan_Status'])
df
```

```python
df.dtypes
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=5)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
X_train_scaled=mms.fit_transform(X_train)
X_test_scaled=mms.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train_scaled,y_train)
```

```python
y_pred=knn.predict(X_test_scaled)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay,classification_report
print(accuracy_score(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
