```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/DATASETS/train-data.csv')
df
```

```python
df.isna().sum()
```

```python
df.drop(['Unnamed: 0','New_Price'],axis=1,inplace=True)
df
```

```python
df.dtypes
```

```python
df['Mileage']=df['Mileage'].str.replace('kmpl','')
df['Mileage']=df['Mileage'].str.replace('km/kg','')
df['Engine']=df['Engine'].str.replace('CC','')
df['Power']=df['Power'].str.replace('bhp','')
df['Power']=df['Power'].str.replace('null','0')
df
```

```python
df.loc[df['Power']=='null '][['Power']]
```

```python
df['Mileage']=df['Mileage'].astype(float)
df['Engine']=df['Engine'].astype(float)
df['Power']=df['Power'].astype(float)
```

```python
df.isna().sum()
```

```python
import numpy as np
```

```python
df.loc[df.Power==0,'Power']=np.NaN
```

```python
lst1=['Engine','Seats']
for i in lst1:
  df[i]=df[i].fillna(df[i].mode()[0])
lst2=['Mileage','Power']
for i in lst2:
  df[i]=df[i].fillna(df[i].mean())
```

```python
df
```

```python
df['brand']=df['Name'].apply(lambda x:x.split()[0])
df
```

```python
df.drop(['Name'],axis=1,inplace=True)
df
```

```python
# labelencoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Owner_Type']=le.fit_transform(df['Owner_Type'])
le1=LabelEncoder()
df['brand']=le1.fit_transform(df['brand'])
le2=LabelEncoder()
df['Transmission']=le2.fit_transform(df['Transmission'])
le3=LabelEncoder()
df['Fuel_Type']=le3.fit_transform(df['Fuel_Type'])
le4=LabelEncoder()
df['Location']=le4.fit_transform(df['Location'])
df
df
```

```python
mapping=list(zip(le.classes_,le.transform(le.classes_)))
mapping
```

```python
df.dtypes
```

```python
X=df.drop(['Price'],axis=1)
X
```

```python
y=df.iloc[:,-2]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(X_train,y_train)
```

```python
y_pred=mlr.predict(X_test)
y_pred
```

```python
y_test
```

```python
import numpy as np
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("r2_score=",r2_score(y_test,y_pred))
print("MAE=",mean_absolute_error(y_test,y_pred))
print("MSE=",mean_squared_error(y_test,y_pred))
print("RMSE=",np.sqrt(mean_squared_error(y_test,y_pred)))
```
